```python
import requests
from bs4 import BeautifulSoup as BS
import numpy as np
import pandas as pd
import re
import statsmodels.api as sm
import statsmodels.graphics as g
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
import time
from jupyterthemes import jtplot
jtplot.style()
```

```python
SOF = pd.read_csv('../data/us_rep_source_of_funds.csv')
```

```python
SOF
```

```python
TOC = SOF.groupby('type of contribution', as_index = False)['amount'].sum()
TOC['proportion'] = TOC['amount']/TOC['amount'].sum()
TOC_party = SOF.groupby(['type of contribution', 'party'], as_index = False)['amount'].sum()
```

```python
TOC.sort_values('amount', ascending = False)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (25,6))

sns.barplot(y = 'type of contribution', 
            x = 'amount', 
            data = TOC,
            orient = 'h',
            order = TOC.sort_values('amount', ascending = False)['type of contribution'],
            ax = ax1
           );
```

```python
TOC_party.sort_values('amount', ascending = False)
```

```python
fig, ax = plt.subplots(figsize = (10,6))
sns.barplot(y = 'type of contribution', 
            x = 'amount', 
            hue = 'party', 
            data = TOC_party[TOC_party['party'].isin(['D','R'])],
            orient = 'h',
            order = TOC.sort_values('amount', ascending = False)['type of contribution'],
            ax = ax
           );
```

```python
TOC_DR = TOC_party[TOC_party['party'].isin(['D','R'])]
TOC_DR_wide = TOC_DR.pivot_table(values = 'amount', index = 'type of contribution', columns = 'party')
```

```python
TOC_DR['party proportion amount'] = TOC_DR['amount']/TOC_DR.groupby('party')['amount'].transform('sum')
fig, ax = plt.subplots(figsize = (10,6))
sns.barplot(y = 'type of contribution', 
            x = 'party proportion amount', 
            hue = 'party', 
            data = TOC_DR,
            orient = 'h',
            order = TOC.sort_values('amount', ascending = False)['type of contribution'],
            ax = ax
           );
```
